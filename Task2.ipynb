```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# read csv file
df = pd.read_csv("train.csv")
df
```

```python
df.info()
```

```python
print(df.isnull().sum())
```

```python
# Impute missing values for 'Age' column with median
df['Age'] = df['Age'].fillna(df['Age'].median())

# Impute missing values for 'Embarked' column with mode
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])

# Drop 'Cabin' column as it has a lot of missing values
df.drop('Cabin', axis=1, inplace=True)

# Check for any remaining missing values
print(df.isnull().sum())
```

```python
# Explore the distribution of 'Survived' column
sns.countplot(x='Survived', data=df)
plt.title('Survival Count')
plt.show()
```

```python
# Explore the relationship between 'Survived' and 'Sex'
sns.countplot(x='Survived', hue='Sex', data=df)
plt.title('Survival by Gender')
plt.show()
```

```python
# Explore the distribution of 'Age' column
sns.histplot(df['Age'], bins=20, kde=True)
plt.title('Age Distribution')
plt.show()
```

```python
# Explore the relationship between 'Survived' and 'Embarked'
sns.countplot(x='Survived', hue='Embarked', data=df)
plt.title('Survival by Embarked')
plt.show()
```

```python
# Scatterplot to determine the relation between 'Age' & 'Fare' for Survival
sns.scatterplot(x = 'Age', y = 'Fare', hue='Survived', data=df)
plt.show()
```
